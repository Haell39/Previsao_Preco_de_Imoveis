## Relatório Sucinto da Matriz de Correlação (Dataset de Casas da Califórnia)
---

 -1 (correlação negativa perfeita)
  1 (correlação positiva perfeita)
  0 (correlação nula)

### **Principais Observações:**

* **`median_income` e `median_house_value`:** Forte correlação **positiva** (0.89). Sugere que, quanto maior a renda média de uma área, maior o valor médio das casas.
* **`total_rooms`, `total_bedrooms`, `population`, `household`:** Estas variáveis estão **fortemente correlacionadas positivamente entre si** (valores próximos a 0.99 ou 1.00). Isso é esperado, pois são contagens relacionadas ao tamanho e densidade das propriedades e comunidades. Pode indicar multicolinearidade se usadas juntas em modelos.
* **`housing_median_age` e `total_rooms`/`total_bedrooms`/`population`/`households`:** Correlação **negativa** notável (cerca de -0.75 a -0.81). Casas mais antigas tendem a ter menos quartos, população, etc., ou estão em áreas com essas características.
* **`latitude` e `longitude`:** Correlação **negativa** (-0.98). Como esperado, essas coordenadas são inversamente correlacionadas, pois à medida que se move para o leste (longitude crescente), a latitude geralmente diminui, e vice-versa, refletindo a geografia da Califórnia.
* **Correlações Fracas/Negligenciáveis:** Muitas outras correlações são próximas de zero, indicando pouca ou nenhuma relação linear direta entre essas variáveis.

### **Conclusão:**

A variável `median_income` é o preditor mais forte do `median_house_value` com base na correlação linear. As contagens de `rooms`, `bedrooms`, `population` e `households` são altamente inter-relacionadas e, juntamente com `housing_median_age`, refletem características estruturais e demográficas do bairro.

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
import pandas as pd
import numpy as np

#  juntando X_test e y_test para aplicar as mesmas transformações que foram feitas em train_data
test_data = X_test.join(y_test)

#  Aplicando log + 1 nas colunas numéricas
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)

#  One-Hot Encoding e removendo a coluna original 'ocean_proximity'
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

# criando as novas features 'bedroom_ratio' e 'households_rooms'
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['households_rooms'] = test_data['total_rooms'] / test_data['households']


trained_feature_columns = x_train.columns

X_test_final = test_data.reindex(columns=trained_feature_columns, fill_value=0)

y_test_final = y_test
```

```python
X_test = X_test_final
y_test = y_test_final
```

```python
reg.score(X_test_final, y_test_final)
```

```python
"""Random Forest"""

from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=100, random_state=27, n_jobs=-1)

forest.fit(x_train, y_train)

print("RandomForestRegressor treinado com sucesso!")
```

```python
score_forest = forest.score(X_test_final, y_test_final)
print(f"R^2 Score do RandomForestRegressor no conjunto de teste: {score_forest}")
```
