### **Principais ObservaÃ§Ãµes:**

* **`median_income` e `median_house_value`:** Forte correlaÃ§Ã£o **positiva** (0.89). Sugere que, quanto maior a renda mÃ©dia de uma Ã¡rea, maior o valor mÃ©dio das casas.
* **`total_rooms`, `total_bedrooms`, `population`, `household`:** Estas variÃ¡veis estÃ£o **fortemente correlacionadas positivamente entre si** (valores prÃ³ximos a 0.99 ou 1.00). Isso Ã© esperado, pois sÃ£o contagens relacionadas ao tamanho e densidade das propriedades e comunidades. Pode indicar multicolinearidade se usadas juntas em modelos.
* **`housing_median_age` e `total_rooms`/`total_bedrooms`/`population`/`households`:** CorrelaÃ§Ã£o **negativa** notÃ¡vel (cerca de -0.75 a -0.81). Casas mais antigas tendem a ter menos quartos, populaÃ§Ã£o, etc., ou estÃ£o em Ã¡reas com essas caracterÃ­sticas.
* **`latitude` e `longitude`:** CorrelaÃ§Ã£o **negativa** (-0.98). Como esperado, essas coordenadas sÃ£o inversamente correlacionadas, pois Ã  medida que se move para o leste (longitude crescente), a latitude geralmente diminui, e vice-versa, refletindo a geografia da CalifÃ³rnia.
* **CorrelaÃ§Ãµes Fracas/NegligenciÃ¡veis:** Muitas outras correlaÃ§Ãµes sÃ£o prÃ³ximas de zero, indicando pouca ou nenhuma relaÃ§Ã£o linear direta entre essas variÃ¡veis.

### **ConclusÃ£o:**

A variÃ¡vel `median_income` Ã© o preditor mais forte do `median_house_value` com base na correlaÃ§Ã£o linear. As contagens de `rooms`, `bedrooms`, `population` e `households` sÃ£o altamente inter-relacionadas e, juntamente com `housing_median_age`, refletem caracterÃ­sticas estruturais e demogrÃ¡ficas do bairro.

```python
# taking the logs
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
```
