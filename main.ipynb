### **Principais ObservaÃ§Ãµes:**

* **`median_income` e `median_house_value`:** Forte correlaÃ§Ã£o **positiva** (0.89). Sugere que, quanto maior a renda mÃ©dia de uma Ã¡rea, maior o valor mÃ©dio das casas.
* **`total_rooms`, `total_bedrooms`, `population`, `household`:** Estas variÃ¡veis estÃ£o **fortemente correlacionadas positivamente entre si** (valores prÃ³ximos a 0.99 ou 1.00). Isso Ã© esperado, pois sÃ£o contagens relacionadas ao tamanho e densidade das propriedades e comunidades. Pode indicar multicolinearidade se usadas juntas em modelos.
* **`housing_median_age` e `total_rooms`/`total_bedrooms`/`population`/`households`:** CorrelaÃ§Ã£o **negativa** notÃ¡vel (cerca de -0.75 a -0.81). Casas mais antigas tendem a ter menos quartos, populaÃ§Ã£o, etc., ou estÃ£o em Ã¡reas com essas caracterÃ­sticas.
* **`latitude` e `longitude`:** CorrelaÃ§Ã£o **negativa** (-0.98). Como esperado, essas coordenadas sÃ£o inversamente correlacionadas, pois Ã  medida que se move para o leste (longitude crescente), a latitude geralmente diminui, e vice-versa, refletindo a geografia da CalifÃ³rnia.
* **CorrelaÃ§Ãµes Fracas/NegligenciÃ¡veis:** Muitas outras correlaÃ§Ãµes sÃ£o prÃ³ximas de zero, indicando pouca ou nenhuma relaÃ§Ã£o linear direta entre essas variÃ¡veis.

### **ConclusÃ£o:**

A variÃ¡vel `median_income` Ã© o preditor mais forte do `median_house_value` com base na correlaÃ§Ã£o linear. As contagens de `rooms`, `bedrooms`, `population` e `households` sÃ£o altamente inter-relacionadas e, juntamente com `housing_median_age`, refletem caracterÃ­sticas estruturais e demogrÃ¡ficas do bairro.

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
import pandas as pd
import numpy as np

#  juntando X_test e y_test para aplicar as mesmas transformaÃ§Ãµes que foram feitas em train_data
test_data = X_test.join(y_test)

#  Aplicando log + 1 nas colunas numÃ©ricas
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)

#  One-Hot Encoding e removendo a coluna original 'ocean_proximity'
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

# criando as novas features 'bedroom_ratio' e 'households_rooms'
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['households_rooms'] = test_data['total_rooms'] / test_data['households']


trained_feature_columns = x_train.columns

X_test_final = test_data.reindex(columns=trained_feature_columns, fill_value=0)

y_test_final = y_test
```

```python
X_test = X_test_final
y_test = y_test_final
```

```python
reg.score(X_test_final, y_test_final)
```

```python
"""Random Forest"""

from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=100, random_state=27, n_jobs=-1)

forest.fit(x_train, y_train)

print("RandomForestRegressor treinado com sucesso!")
```

```python
score_forest = forest.score(X_test_final, y_test_final)
print(f"Score do RandomForestRegressor no conjunto de teste: {score_forest}")
```

```python
"""Cross Validation"""
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()
param_grid = {'n_estimators': [30, 50, 100], "max_features": [8, 12, 20], "min_samples_split": [2, 4, 6, 8]}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring = "neg_mean_squared_error", return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test, y_test)
```
